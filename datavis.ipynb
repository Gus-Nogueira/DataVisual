```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x = np.linspace(0, 20, 20)
y=[math.sqrt(i) for i in x]

plt.rcParams.update({'font.size': 16})
plt.figure(figsize = (8,5))

plt.plot(x,y, 'gv', label = 'Raíz')
plt.plot([1,2,3,4,5,6,7], color = 'crimson')
plt.xlabel('Eixo X', fontsize = 16, fontstyle = 'italic')
plt.ylabel('Eixo Y', fontsize = 16, fontstyle = 'italic')
plt.title("Gráfico simples", fontsize = 18, fontstyle = 'italic', fontweight = 'bold')
plt.xlim([0,6])
plt.ylim([0,4])
plt.savefig('Curva.png', dpi = 600, transparent = True)
plt.legend(loc = 1)
plt.show()
```

```python
x = np.linspace(0, 20, 20)
y=[math.sqrt(i) for i in x]
y1= x**3
plt.rcParams.update({'font.size': 16})
plt.figure(figsize = (8,5))

plt.plot(x, y, 'gv', label = 'Raiz')
plt.plot(x, y1, 'ro', label = 'Cubo')

plt.xlabel('Eixo X', fontsize = 16, fontstyle = 'italic')
plt.ylabel('Eixo Y', fontsize = 16, fontstyle = 'italic')
plt.title("Gráfico das funções", fontsize = 18, fontstyle = 'italic', fontweight = 'bold')
plt.savefig('Curva2.png', dpi = 600, transparent = True)
plt.legend(loc = 0)
plt.show()
```
