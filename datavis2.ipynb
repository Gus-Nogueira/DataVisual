```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x=[1,2,3,4]
y=[1,4,2,3]
a=[4,1,5,3]
b=[1,5,3,4]

plt.plot(x,y, color ='m', linestyle = (0,(3,1,1,1)), label = "magenta")
plt.plot(a,b, color = 'darkorange', linestyle = 'dashdot', label = "laranja")
plt.title("Duas linhas", fontsize = 18, fontweight = 'bold')
plt.xlabel("Horizontal", fontsize = 16)
plt.ylabel("Vertical", fontsize = 16)
plt.legend(loc = 1)
plt.show
```

```python
x=[1,3,5,7,9,11]
y=[5,9,7,3,8,2]

x2=[2,4,6,8,10,12]
y2=[8,5,2,3,9,7]

plt.bar(x,y, color = 'red', label = "barra vermelha")
plt.bar(x2,y2, color = 'm', label = "barra magenta")
plt.title("Barras", fontsize =18, fontweight= 'bold')
plt.xlabel("Horizontal", fontsize =16)
plt.ylabel("Vertical", fontsize =16)
plt.legend(loc = 1)

plt.show()
```

```python
x=[1,3,5,7,9,11]
y=[5,9,7,3,8,2]

x2=[1,3,5,7,9,11]
y2=[5,9,7,3,8,2]

plt.scatter(x,y, color = 'black', marker ='P', label = "pontos pretos")
plt.plot(x,y, color = 'red', linestyle= 'dashdot', label = "linhas vermelhas")
plt.title("Pontos", fontsize =18, fontweight= 'bold')
plt.xlabel("Horizontal", fontsize =16)
plt.ylabel("Vertical", fontsize =16)
plt.legend(loc = 1)
plt.show()
```

```python
plt.rcParams.update({'font.size': 20})

x=[1,3,6]
rotulos=['Senior','Pleno', 'Junior']

fatia=[0,0,0.2]

plt.title("Proporção", fontsize=18, fontweight ='bold')
plt.axis('equal')
plt.pie(x, labels = rotulos, autopct = '%1.1f%%', startangle =90, shadow = True,
        explode=fatia)

plt.show()
```
